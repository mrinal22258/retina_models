### ganglion_cell_data

```python
# 15-10-07

import h5py

file_path = r'C:\Users\nisha\Downloads\neural_code_data\neural_code_data\ganglion_cell_data\15-10-07\naturalscene.h5'

def explore_h5(file):
    def explore_group(group, prefix=''):
        for key in group.keys():
            item = group[key]
            if isinstance(item, h5py.Group):  # If it's a Group
                print(f"Group: {prefix}/{key}")
                explore_group(item, prefix=f"{prefix}/{key}")  # Recursive call
            elif isinstance(item, h5py.Dataset):  # If it's a Dataset
                print(f"Dataset: {prefix}/{key} | Shape: {item.shape}, Dtype: {item.dtype}")

    with h5py.File(file, 'r') as f:
        print("Root Keys:", list(f.keys()))
        explore_group(f)

explore_h5(file_path)
```

```python
import matplotlib.pyplot as plt

file_path = r'C:\Users\nisha\Downloads\neural_code_data\neural_code_data\ganglion_cell_data\15-10-07\naturalscene.h5'

with h5py.File(file_path, 'r') as f:
    stimulus_data = f['/test/stimulus']
    
    image_index = 1
    image = stimulus_data[image_index]  # Shape (50, 50)
    
    plt.imshow(image, cmap='gray')
    plt.title(f"Stimulus Image at Index {image_index}")
    plt.colorbar(label="Pixel Intensity")
    plt.show()
```

```python
with h5py.File(file_path, 'r') as f:
    spikes_group = f['/spikes']
    fig, ax = plt.subplots(figsize=(10, 6))
    
    for i, cell_name in enumerate(spikes_group.keys()):
        spike_times = spikes_group[cell_name][:]
        ax.scatter(spike_times, [i+1]*len(spike_times), s=1, label=cell_name)  # Scatter plot for each cell

    ax.set_title("Raster Plot of Spikes")
    ax.set_xlabel("Time (s)")
    ax.set_ylabel("Cell")
    ax.legend(loc='upper right', fontsize='small')
    plt.show()
```
